```python
"""
Alpha Vantage US options backfill: pulls full option chains by trading day.

- Saves one CSV per day per symbol: data/options/<SYMBOL>/YYYY/MM/DD.csv
- Resumable: existing files are skipped.
- Gentle pacing + retry for "Note" / rate-limit responses.
- Change SYMBOLS, START, END, API_KEY as needed.

Requires: requests, pandas, python-dateutil
"""

import os, time, json, gzip, io
from datetime import date, datetime
from dateutil import tz
import requests
import pandas as pd

API_KEY   = os.getenv("ALPHAVANTAGE_API_KEY", "Q3EBFXC1GEUITND1")
BASE_URL  = "https://www.alphavantage.co/query"
# --- add these imports at top if missing ---
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay



def day_path(symbol, d):
    return os.path.join(OUT_DIR, symbol, f"{d.year:04d}", f"{d.month:02d}", f"{d.day:02d}.csv")

# Try to use a real NYSE calendar if available; otherwise fall back to US federal holidays.
def trading_days(start, end):
    """
    Return actual NYSE trading sessions (preferred) or business days minus US federal holidays (fallback).
    """
    try:
        import pandas_market_calendars as mcal  # pip install pandas-market-calendars
        cal = mcal.get_calendar('XNYS')
        # valid_days returns session dates (midnight tz-aware); strip tz to match the rest of the code
        days = cal.valid_days(start_date=str(start.date()), end_date=str(end.date()), tz='America/New_York')
        return days.tz_convert(None)
    except Exception:
        # Fallback: Mon–Fri minus US federal holidays (close but not perfect for NYSE)
        usb = CustomBusinessDay(calendar=USFederalHolidayCalendar())
        return pd.date_range(start, end, freq=usb)

def symbol_for_date(symbol: str, day: pd.Timestamp) -> str:
    """
    Handle ticker renames where the data provider might require the legacy symbol.
    Alpha Vantage often accepts the modern symbol historically, but this makes it robust.
    """
    if symbol.upper() == "META" and day < pd.Timestamp("2022-06-09"):
        return "FB"
    return symbol


def annotate_df(df: pd.DataFrame, day: pd.Timestamp, requested_symbol: str) -> pd.DataFrame:
    """
    Add non-colliding metadata columns.
    - 'as_of_date' (safe; create or overwrite)
    - '_requested_symbol' (what you asked for)
    - '_api_symbol' (what was actually queried; e.g., FB pre-2022-06-09)
    """
    api_sym = symbol_for_date(requested_symbol, day)
    if 'as_of_date' not in df.columns:
        df.insert(0, 'as_of_date', day.date())
    else:
        df['as_of_date'] = day.date()
    df['_requested_symbol'] = requested_symbol
    df['_api_symbol'] = api_sym
    return df


def fetch_chain(symbol, day, datatype="csv"):
    """Alpha Vantage HISTORICAL_OPTIONS → CSV text or None if 'No data' (holiday/pre-IPO/etc.)."""
    q_symbol = symbol_for_date(symbol, day)
    params = {
        "function": "HISTORICAL_OPTIONS",
        "symbol": q_symbol,
        "date": day.strftime("%Y-%m-%d"),
        "datatype": datatype,
        "apikey": API_KEY,
    }
    headers = {"Accept-Encoding": "gzip"}
    r = requests.get(BASE_URL, params=params, headers=headers, timeout=60)
    if r.status_code != 200:
        raise RuntimeError(f"HTTP {r.status_code} for {symbol} {day.date()}: {r.text[:200]}")
    ctype = r.headers.get("Content-Type","").lower()
    text = r.text.strip()

    # JSON -> either rate limit ('Note'), 'No data', or other error
    if "application/json" in ctype or text.startswith("{"):
        try:
            msg = r.json()
        except Exception:
            raise RuntimeError(f"Unexpected JSON/text for {symbol} {day.date()}: {text[:200]}")
        # Rate limit "Note" → tell caller to retry
        if "Note" in msg:
            raise RuntimeError("RATE_LIMIT: " + msg.get("Note", str(msg)))
        # Clean 'No data' case → tell caller to skip without retry
        if str(msg.get("message", "")).lower().startswith("no data for symbol"):
            return None
        # Other JSON → treat as hard error
        raise RuntimeError(f"API_ERROR: {msg}")

    return r.text  # CSV

def backfill_symbol(symbol, start, end):
    days = trading_days(start, end)
    print(f"[{symbol}] target trading days: {len(days)}")
    for d in days:
        out_fp = day_path(symbol, d)
        if os.path.exists(out_fp):
            continue
        ensure_dir(out_fp)
        for attempt in range(1, MAX_RETRY+1):
            try:
                csv_text = fetch_chain(symbol, d)
                if csv_text is None:
                    # Holiday/pre-IPO/etc. → write a tiny placeholder and move on without retrying
                    with open(out_fp, "w", encoding="utf-8") as f:
                        f.write("empty\n")
                    print(f"[{symbol}] {d.date()} NYSE closed or no dataset yet → skipped")
                    break
                # Empty CSV (edge case) → still write placeholder
                if len(csv_text.splitlines()) <= 1:
                    with open(out_fp, "w", encoding="utf-8") as f:
                        f.write("empty\n")
                    print(f"[{symbol}] {d.date()} empty CSV → skipped")
                    break
                # Normal path: save CSV with annotations
                df = pd.read_csv(io.StringIO(csv_text))
                df = annotate_df(df, d, symbol)
                df.to_csv(out_fp, index=False)
                df.to_csv(out_fp, index=False)
                break
            except Exception as e:
                msg = str(e)
                # Only back off on true rate limits; otherwise bubble up or continue
                if "RATE_LIMIT:" in msg and attempt < MAX_RETRY:
                    sleep_s = PAUSE_S * attempt
                    print(f"[{symbol}] {d.date()} rate limit → retry {attempt}/{MAX_RETRY} in {sleep_s:.1f}s")
                    time.sleep(sleep_s)
                    continue
                if "RATE_LIMIT:" in msg and attempt == MAX_RETRY:
                    print(f"[{symbol}] {d.date()} giving up after rate limits: {msg}")
                    break
                # Non-rate-limit error → log and move on
                print(f"[{symbol}] {d.date()} error (no retry): {msg}")
                break
        time.sleep(PAUSE_S)
```

```python
def ensure_dir(path):
    os.makedirs(os.path.dirname(path), exist_ok=True)
```

```python
SYMBOLS   = ["MSFT"]
START     = pd.Timestamp("2008-01-23")   # earliest supported by HISTORICAL_OPTIONS
END       = pd.Timestamp(date.today())   # today
OUT_DIR   = "data/options"
PAUSE_S   = 12.0  # be kind; adjust for your plan
MAX_RETRY = 3
```

```python
for sym in SYMBOLS:
    backfill_symbol(sym, START, END)
print("Done.")
```
