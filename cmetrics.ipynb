```python
import pandas as pd
import io
import requests

# Coin Metrics hosts community data CSVs on GitHub
# Base URL for the legacy community data repo
base_url = "https://raw.githubusercontent.com/coinmetrics/data/master/csv/{}.csv"

def get_full_history(asset):
    url = base_url.format(asset)
    try:
        s = requests.get(url).content
        df = pd.read_csv(io.StringIO(s.decode('utf-8')))
        
        # Convert date to datetime object
        df['time'] = pd.to_datetime(df['time'])
        
        # Filter for just the price (ReferenceRateUSD) or other metrics you need
        # Note: Column names in CSV might differ slightly from API (e.g., 'PriceUSD' vs 'ReferenceRateUSD')
        return df
    except Exception as e:
        print(f"Error loading data for {asset}: {e}")
        return None

# Get full history for Bitcoin
btc_data = get_full_history("btc")

if btc_data is not None:
    print(f"Loaded {len(btc_data)} days of data!")
    print(btc_data[['time', 'PriceUSD']].tail()) # 'PriceUSD' is often the column name in these CSVs
```
